Tokenization, also known as word segmentation, is the task of splitting a stream of characters into a ordered list of semantically useful groups of characters such as words and punctuation (e.g. the sentence "The cat sat on the mat." consists of tokens "The", "cat", "sat", "on", "the", "mat", "."). For the purpose of this task, you can assume that a token should be separated whenever a space (" ") or punctuation (".", ",", ";", "!") is encountered. Tokenization is a vital part of the NLP pipeline as it enables downstream linguistic analysis of text (e.g. how many nouns are in the text ?).

##### a) Implement the function `word_tokenize` to tokenize an input string. Apply it to the above tweet and print the tokenized result
<i>Hint: Make sure that punctuations like '.', '!' and ';' are returned as separate tokens! </i>

```python
from collections.abc import Mapping
```

```python
tweet = "    While we're waiting for @LeagueOfLegends's 10th anniversary. Let's have some fun! Tweet me your favorite skin ideas and I'll send 1k RP to a couple of my favorites. My friend keeps telling me he'd like to see ShamWow! Vlad. #League10    "

def word_tokenize(text):
    pun=[".",".","!",";"]
    string=' ' 
    for ch in (text):
        if ch in pun:
            string=string+' '+ch+' '  # add space before and after the specific punctuation in the text
        else:
            string+=ch
    string=string.split()            #  via space tokenize                
    return string

token_s=word_tokenize(tweet)           
print(token_s)

   
```

##### b) Print all tokens and their length. Each token must only be printed once.

```python
clean_tokens=set(token_s)  # remove duplicates
token_length={}
for word in clean_tokens:
    token_length[word]=len(word)  # get length of each word and save in dictionary
print(token_length)
```

##### c) Print all tokens and their absoulte frequency. Each token must only be printed once.

```python
clean_tokens=list(clean_tokens)
freq={}
for i in range(len(clean_tokens)):
    count=0
    for j in range(len(token_s)):          # calculate the frequency of each token
        if token_s[j]==clean_tokens[i]:
            count+=1
    freq[clean_tokens[i]]=count            # the results save in dictionary    
print(freq)    
```

##### b) Implement the method `count_bigrams`. It shall calculate bigram counts for a series of tokenized texts. (Bigrams are sequences of two tokens).
__Hint__: You are allowed to use standard python libararies. Checkout the package <a href="https://docs.python.org/3/library/collections.html#module-collections">collections</a> and the suggested return type which may make your life easier. <br>
Example: Consider the two sentences "I like pizza" and "I like pasta" <br>
Then we have the following bigrams: (I, like), (like, pizza), (I,like), (like, pasta). The function `count_bigrams` should return which bigrams exist and how often they occur. In the example the result would be something like: <br>

|          |      | 
|--------------|-----------|
| (I, like) | 2      |
| (like, pizza) | 1      |
| (like, pasta) | 1      |


```python
def count_bigrams(tokenized_texts): 
    cnt=Counter()
    bigrams=[]
    for i in range(len(tokenized_texts)):
        string=tokenized_texts[i]  

        for j in range(len(string)-1):
            bg=(string[j],string[j+1])
            bigrams.append(bg)              # bigram
    cnt=Counter(bigrams)                    # count the bigrams
    return cnt
 

tokenized_texts=df['result']
freq={}
freq=count_bigrams(tokenized_texts)
print(DataFrame(freq.most_common(len(freq)),columns=['bigrams', 'cnt']))
```

##### b) Implement the method `get_probability_of_bigrams`.  It shall calculate the probability of the top 20 frequent bigrams (for positive and negative) from 2a). Follow the definition stated above and compute the probabilities based on the full dataset provided (i.e. based on all bigrams in all reviews).
Example: If there are 100 bigrams in total and the bigram ("a", "b") occurs 5 times in positive documents, your result should be 0.05 for positive documents.

```python
def get_probability_of_bigrams(sum_bigrams, top_k):
    probability=pd.DataFrame(columns=['bigram','probability'])    #create dataframe
    for bigram in top_k:
        line={'bigram':bigram[0],'probability':bigram[1]/sum_bigrams}    #count probability
        
        probability=probability.append(line,ignore_index=True)
    return probability
    

    
total_bigrams=set(freq.elements())
count=0
for bigram in total_bigrams:         # count the total number of bigrams
    count+=freq[bigram]

pos_list=get_top_k_bigrams(data_frame,1,20)  #get the most frequent positive bigrams and count
neg_list=get_top_k_bigrams(data_frame,0,20)  #get the most frequent negative bigrams and count

p_big_pos=get_probability_of_bigrams(count, pos_list)
p_big_neg=get_probability_of_bigrams(count, neg_list)
print("the probability of the top 20 frequent bigrams for positive:",'\n',p_big_pos)
print("the probability of the top 20 frequent bigrams for negative:",'\n',p_big_neg)
```
